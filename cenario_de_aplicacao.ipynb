## **Problema, Contexto e Justificativa**
A administração correta de medicamentos é uma atividade essencial na rotina de pessoas idosas, especialmente aquelas que fazem uso contínuo de múltiplos comprimidos ao longo do dia. No entanto, fatores como limitações visuais, dificuldades motoras e lapsos de atenção tornam a contagem manual de comprimidos uma tarefa suscetível a erros.

Esses erros podem resultar na ingestão incorreta de medicamentos, como doses duplicadas ou esquecidas, representando riscos significativos à saúde. Além disso, muitos idosos dependem do auxílio de terceiros para realizar essa tarefa, o que reduz sua autonomia e independência no dia a dia.

Diante desse contexto, torna-se relevante o desenvolvimento de uma solução tecnológica simples, acessível e de fácil utilização, capaz de auxiliar na contagem precisa de comprimidos, promovendo maior segurança e autonomia para o usuário idoso.

## **Objetivo do Sistema**
O objetivo do sistema é desenvolver uma aplicação de processamento visual que permita a identificação e contagem automática de comprimidos em tempo real, utilizando uma webcam.

O sistema deverá fornecer ao usuário idoso uma interface simples e intuitiva, apresentando de forma clara a quantidade de comprimidos detectados, além de permitir a comparação com uma quantidade esperada, auxiliando na verificação correta da medicação antes do consumo.

## **Funcionamento e Interatividade**
O sistema será operado de maneira simples, visando atender usuários com pouca familiaridade com tecnologia. O usuário deverá posicionar os comprimidos sobre uma superfície plana e bem iluminada, visível para a câmera.

A aplicação realizará automaticamente a captura e o processamento da imagem, identificando os comprimidos presentes na cena e exibindo os resultados em tempo real.

## **Benefícios ao Usuário**
1. Maior autonomia para pessoas idosas na organização de seus medicamentos
2. Redução de erros na contagem manual de comprimidos
3. Aumento da segurança no uso de medicamentos
4. Interface simples e acessível, adequada para usuários leigos
